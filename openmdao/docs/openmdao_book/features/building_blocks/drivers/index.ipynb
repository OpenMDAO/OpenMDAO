# Drivers

Drivers repeatedly interrogate an OpenMDAO model.  Generally they serve one of two purposes: optimization or design space exploration.

## Optimization drivers

- [ScipyOptimizeDriver](scipy_optimize_driver.ipynb)
- [pyOptSparseDriver](pyoptsparse_driver.ipynb)
- [SimpleGADriver](genetic_algorithm.ipynb)
- [DifferentialEvolutionDriver](differential_evolution.ipynb)

## Design space exploration drivers

- [AnalysisDriver](analysis_driver.ipynb)
- [DOEDriver](doe_driver.ipynb)

## Autoscalers

While some optimizers provide automatic scaling of the optimization problem, OpenMDAO itself also includes an automatic scaling capability in addition to allowing users to manually specify scaling.

- [Autoscaler API](autoscaler_api.ipynb)
